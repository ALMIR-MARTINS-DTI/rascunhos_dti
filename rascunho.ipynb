```python
def configura_excel(df):
    """
    Configura o DataFrame para salvar em Excel com as seguintes regras:
    - Nas colunas booleanas (receptor_estrogeno.acertou, receptor_progesterona.acertou, status_her2.acertou, status_ck5_6.acertou, ki67_percentual.acertou), colorir de vermelho se False.
    - Na coluna (erro), colorir de vermelho se True.
    - Salvar em uma aba chamada 'Resultados' e outra aba chamada 'Resumo' com a contagem de erros e o percentual de erros por coluna em relação ao total de linhas, inserir uma linha com o total de erros e do percentual.
    - Salvar o arquivo como 'resultados_imuno_formatados.xlsx'.
    """

    import pandas as pd

    # Função para aplicar a formatação condicional
    def colorir_vermelho(val):
        color = 'red' if val == False else ''
        return f'background-color: {color}'
    
    # def colorir_verde(val):
    #     color = 'green' if val == True else ''
    #     return f'background-color: {color}'

    # Seleciona as colunas booleanas para aplicar a formatação
    colunas_booleanas = [
        'receptor_estrogeno.acertou',
        'receptor_progesterona.acertou',
        'status_her2.acertou',
        'status_ck5_6.acertou',
        'ki67_percentual.acertou',
    ]

    # Cria um objeto Styler
    styler = df.style

    # Aplica a formatação condicional nas colunas booleanas
    for coluna in colunas_booleanas:
        if coluna in df.columns:
            styler = styler.map(colorir_vermelho, subset=[coluna])
    
    # Aplica a formatação condicional na coluna 'erro'
    if 'erro' in df.columns:
        styler = styler.map(colorir_vermelho, subset=['erro'])

    # Cria o resumo de erros por coluna
    styler = styler.map(colorir_vermelho, subset=[coluna])

    # Cria o resumo de erros por coluna
    resumo = {coluna: df[coluna].value_counts().get(False, 0) for coluna in colunas_booleanas}
    resumo_df = pd.DataFrame(list(resumo.items()), columns=['Coluna', 'Contagem de Erros'])
    resumo_df['Percentual de Erros (%)'] = (resumo_df['Contagem de Erros'] / len(df)) * 100
    resumo_df.loc['Total'] = resumo_df.sum(numeric_only=True)

    # Salva em um arquivo Excel com duas abas
    with pd.ExcelWriter('resultados_imuno_formatados.xlsx', engine='openpyxl') as writer:
        styler.to_excel(writer, sheet_name='Resultados', index=False)
        resumo_df.to_excel(writer, sheet_name='Resumo', index=False)

    print("Arquivo Excel salvo como 'resultados_imuno_formatados.xlsx' com as abas 'Resultados' e 'Resumo'.")
    display(resumo_df.head(1))

# configura_excel(df_laura)
```

```python
import re

def extrai_status_ck5_6(txt: str) -> str:
    """
    Extrai o status do CK5/6: "POSITIVO", "NEGATIVO" ou retorna "NÃO INFORMADO".
    Esta versão mantém a verificação de distância e resolve o problema de ambiguidade.
    """
    txt_lower = txt.lower()

    # 1. Verificar abreviações (CK5/6+ ou CK5/6-) - Estes são os mais específicos e diretos.
    # Mantemos a busca no 'txt' original com IGNORECASE para estas abreviações.
    if re.search(r"\bck5\s*\/\s*6\s*[\+]\b", txt, flags=re.IGNORECASE):
        return "POSITIVO"
    if re.search(r"\bck5\s*\/\s*6\s*[\-]\b", txt, flags=re.IGNORECASE):
        return "NEGATIVO"

    # 2. Procurar por "CK5/6" e o status mais próximo a ele.
    # max_chars_after_ck5_6 é a sua verificação de distância!
    # Ela define o quão longe o regex pode "olhar" após encontrar "CK5/6".
    max_chars_after_ck5_6 = 70 # Ajuste este valor conforme a variabilidade dos seus relatórios

    # O padrão busca "ck5/6", então *qualquer coisa* (não-ganancioso)
    # até o limite de 'max_chars_after_ck5_6', e então o *primeiro* "negativo" ou "positivo".
    match = re.search(
        r"ck5\s*\/\s*6.{0," + str(max_chars_after_ck5_6) + r"}?(\bnegativo\b|\bpositivo\b)",
        txt_lower
    )

    if match:
        status_found = match.group(1) # Captura o que foi matched no primeiro grupo (negativo ou positivo)
        if status_found == "positivo":
            return "POSITIVO"
        elif status_found == "negativo":
            return "NEGATIVO"

    # Se nenhuma das condições acima foi atendida
    return "NÃO INFORMADO"


# Teste da função
texto = """
MATERIAL ID:43 MATERIAL - Biópsias de mama esquerda - QSM (PA) e mama direita - UQL (PB) RE - RECEPTOR DE ESTRÓGENO - PA - POSITIVO COM INTENSIDADE VARIÁVEL EM 50% DAS CÉLULAS EPITELIAIS DE INTERESSE. Anti-Receptor de Estrógeno / Clone SP1 / Rabbit Monoclonal / Ventana RP - RECEPTOR DE PROGESTERONA - PA - POSITIVO COM INTENSIDADE VARIÁVEL EM 30% DAS CÉLULAS EPITELIAIS DE INTERESSE. Anti-Receptor de Progesterona / Clone 1E2 / Rabbit Monoclonal / Ventana HER-2 - PA - NEGATIVO Anti-HER-2/neu / Clone 4B5 / Rabbit Monoclonal / Ventana KI67 - PA - POSITIVO EM 2% DAS CÉLULAS EPITELIAIS DE INTERESSE. Anti-KI-67 - Clone MIB-1/ Monoclonal/ Dako P63 - PA - POSITIVO AO REDOR DOS DUCTOS E ÁCINOS. Anti-Proteina P63 / Clone DAK-p63 / Monoclonal / DAKO RE - RECEPTOR DE ESTRÓGENO - PB - POSITIVO COM INTENSIDADE MODERADA A FORTE EM 95% DAS CÉLULAS NEOPLÁSICAS Anti-Receptor de Estrógeno / Clone SP1 / Rabbit Monoclonal / Ventana RP - RECEPTOR DE PROGESTERONA - PB - POSITIVO COM INTENSIDADE MODERADA EM 80% DAS CÉLULAS NEOPLÁSICAS Anti-Receptor de Progesterona / Clone 1E2 / Rabbit Monoclonal / Ventana HER-2 - PB - ESCORE 1+ (IMUNOMARCAÇÃO INCOMPLETA DA MEMBRANA, FRACA, POUCO PERCEPTÍVEL, DETECTADA EM 30% DAS CÉLULAS NEOPLÁSICAS). Anti-HER-2/neu / Clone 4B5 / Rabbit Monoclonal / Ventana KI67 - PB - POSITIVO EM 8% DAS CÉLULAS NEOPLÁSICAS Anti-KI-67 - Clone MIB-1/ Monoclonal/ Dako P63 - PB - POSITIVO AO REDOR DOS DUCTOS E ÁCINOS. Anti-Proteina P63 / Clone DAK-p63 / Monoclonal / DAKO CK5/6 - PB - NEGATIVO NAS CÉLULAS NEOPLÁSICAS Anti-Citoqueratina 5/6 / Clone D5/16B4 / Monoclonal / Ventana Reações de imunoistoquímica realizadas com os kits de detecção Ventana Ultraview Universal DAB e/ou Ventana Optiview Universal DAB e/ou Sistema EnvisionFlex HRP-Dako. INTERPRETAÇÃO PA) Biópsia percutânea à vácuo (mamotomia) de mama esquerda - QSM: Parênquima mamário com:. Lesão epitelial proliferativa sem atipias com: - Alterações fibrocísticas com metaplasia apócrina. - Adenose simples. - Adenose de ductos rombos ("blunt duct adenosis"). - Alterações colunares dos lóbulos. - Fibrose e hialinização estromal. - Infiltração adiposa do estroma. - Calcificações basofílicas luminais múltiplas. - Ausência de evidências histológicas de malignidade nesta amostra. PB) Biópsia percutânea à vácuo (mamotomia) de mama direita - UQL: - Carcinoma ductal "in situ" de baixo grau (ver nota). - Não foram caracterizados focos de invasão estromal nos cortes avaliados (ver nota). Nota: O diagnóstico final depende da análise histológica completa da lesão (B). Análise morfológica do material realizada por patologia digital através do sistema de escaneamento de lâminas Aperio GT450 - Leica Biosystems.. Controles internos presentes e adequados nos cortes das reações dos receptores hormonais (estrógeno e progesterona). Her-2 (CERBB2) - Interpretação (segundo critérios do CAP - College of American Pathologists) Detecção por método imunoistoquímico da oncoproteína HER-2. A reação pode sofrer a interferência do processo de fixação prévia. Estes testes não foram validados para material submetido a descalcificação. Valores de referência: ESCORE 0 - Ausência de imunomarcação nas células neoplásicas ou imunomarcação incompleta da membrana, fraca, pouco perceptível, detectada em menos de 10% das células neoplásicas ESCORE 1+ - Imunomarcação incompleta da membrana, fraca, pouco perceptível, detectada em mais de 10% das células neoplásicas ESCORE 2+ (EQUÍVOCO) - Imunomarcação completa da membrana, com intensidade fraca a moderada, em mais de 10% das células neoplásicas ou imunomarcação completa da membrana, com forte intensidade, em menos de 10% das células neoplásicas ESCORE 3+ (POSITIVO) - Imunomarcação completa da membrana, com forte intensidade, em mais de 10% das células neoplásicas Referência: Wolff AC, Somerfield MR, Dowsett M, et al. Human Epidermal Growth Factor Receptor 2 Testing in Breast Cancer: ASCO-College of American Pathologists Guideline Update. J Clin Oncol. 2023: JCO2202864. doi: 10.1200/JCO.22.02864.
"""
print(extrai_status_ck5_6(texto))
print()
conta_siglas(texto)
```

```python
# CORREÇÕES DAS FUNÇÕES PÓS VALIDAÇÃO (LAURA)
import re

def extrai_status_her2(txt: str) -> str:
    """
    Extrai o Status do HER-2 baseado em termos de score:
    - "HER-2 - ESCORE 0" ou "1+" → "NEGATIVO"
    - "HER-2 - ESCORE 2+" → "INCONCLUSIVO"
    - "HER-2 - ESCORE 3+" → "POSITIVO"
    - Novos formatos: "HER-2 - POSITIVO", "HER-2 - NEGATIVO"
    Ajustado para evitar a captura de scores de "Valores de referência" distantes,
    priorizando a extração do score da amostra.
    """
    txt_lower = txt.lower()

    # Define um limite de caracteres para a busca do status após "HER-2".
    # Este valor deve ser ajustado com base na variabilidade dos seus relatórios.
    # Deve ser suficiente para cobrir um resultado direto (ex: "HER-2 - ESCORE 0 (STATUS DE HER2 - NEGATIVO)")
    # e pequeno o suficiente para não pular para seções de referência distantes.
    max_chars_between_her2_and_status = 60 # Mantido o mesmo valor para consistência.

    # 1. BUSCA PADRÃO "HER-2 ... ESCORE X+" (PRIORIDADE MAIOR)
    # Este padrão é mais específico para o formato de resultado da amostra.
    m = re.search(
        r"her[-\s]?2.{0," + str(max_chars_between_her2_and_status) + r"}?s?core\s*[:\-]?\s*([0-3]\+?)",
        txt_lower,
        flags=re.IGNORECASE
    )
    if m:
        score = m.group(1)
        if score.startswith("0") or score == "1+":
            return "NEGATIVO"
        if score == "2+":
            return "INCONCLUSIVO"
        if score == "3+":
            return "POSITIVO"

    # 2. Padrão para "HER-2 ... POSITIVO" (AGORA COM LIMITE DE DISTÂNCIA E PRIORIDADE MENOR)
    # Este é verificado APÓS o padrão de escore.
    if re.search(
        r"her[-\s]?2.{0," + str(max_chars_between_her2_and_status) + r"}?[:\-]?\s*positivo\b",
        txt_lower,
        flags=re.IGNORECASE
    ):
        return "POSITIVO"

    # 3. Padrão para "HER-2 ... NEGATIVO" (AGORA COM LIMITE DE DISTÂNCIA E PRIORIDADE MENOR)
    # Este é verificado APÓS o padrão de escore.
    if re.search(
        r"her[-\s]?2.{0," + str(max_chars_between_her2_and_status) + r"}?[:\-]?\s*negativo\b",
        txt_lower,
        flags=re.IGNORECASE
    ):
        return "NEGATIVO"

    # 4. Outra forma: "her2 3+" isolado (PRIORIDADE AINDA MENOR)
    # Este padrão é mais restrito por não ter um `.*?` e por exigir proximidade.
    m2 = re.search(r"\bher[-]?2\s*[:\-]?\s*([0-3]\+)\b", txt_lower, flags=re.IGNORECASE)
    if m2:
        score2 = m2.group(1)
        if score2 == "1+":
            return "NEGATIVO"
        if score2 == "2+":
            return "INCONCLUSIVO"
        if score2 == "3+":
            return "POSITIVO"

    return "NÃO INFORMADO"


TEXTO3 = """
EXAME IMUNOISTOQUÍMICO MATERIAL ID:15 MATERIAL - Biópsia de mama direita - QIM, distorção arquitetural (A) e calcificações (B) DESCRIÇÃO - O exame imunoistoquimico evidenciou os seguintes resultados: RE - RECEPTOR DE ESTRÓGENO - POSITIVO COM INTENSIDADE MODERADA EM 90% DAS CÉLULAS EPITELIAIS ATÍPICAS (FOCO DE HLA) (A). Anti-Receptor de Estrógeno / Clone SP1 / Rabbit Monoclonal / Ventana RP - RECEPTOR DE PROGESTERONA - POSITIVO COM INTENSIDADE MODERADA EM 90% DAS CÉLULAS EPITELIAIS ATÍPICAS (FOCO DE HLA) (A). Anti-Receptor de Progesterona / Clone 1E2 / Rabbit Monoclonal / Ventana ECADE - NEGATIVO NAS CÉLULAS EPITELIAIS ATÍPICAS (FOCO DE HLA) (A). E-Cadherin / Clone NCH-38 / Dako P63 - POSITIVO AO REDOR DOS DUCTOS E ÁCINOS (A). Anti-Proteina P63 / Clone DAK-p63 / Monoclonal / DAKO RE - RECEPTOR DE ESTRÓGENO - POSITIVO COM INTENSIDADE MODERADA EM 70% DAS CÉLULAS NEOPLÁSICAS (B). Anti-Receptor de Estrógeno / Clone SP1 / Rabbit Monoclonal / Ventana RP - RECEPTOR DE PROGESTERONA - NEGATIVO NAS CÉLULAS NEOPLÁSICAS (B). Anti-Receptor de Progesterona / Clone 1E2 / Rabbit Monoclonal / Ventana ECADE - POSITIVO NAS CÉLULAS NEOPLÁSICAS (B). E-Cadherin / Clone NCH-38 / Dako CK5/6 - NEGATIVO NAS CÉLULAS NEOPLÁSICAS (B). Anti-Citoqueratina 5/6 / Clone D5/16B4 / Monoclonal / Ventana HER-2 - ESCORE 3+ (STATUS DE HER2 - POSITIVO) - IMUNOMARCAÇÃO FORTE E COMPLETA, DESENHANDO TODO O CONTORNO DA MEMBRANA EM 90% DAS CÉLULAS NEOPLÁSICAS (B). Anti-HER-2/neu / Clone 4B5 / Rabbit Monoclonal / Ventana KI67 - POSITIVO EM 40% DAS CÉLULAS NEOPLÁSICAS (B). Anti-KI-67 - Clone MIB-1/ Monoclonal/ Dako P63 - POSITIVO (CONTÍNUO E DESCONTÍNUO) AO REDOR DOS DUCTOS E ÁCINOS, NOS CORTES AVALIADOS (B). Anti-Proteina P63 / Clone DAK-p63 / Monoclonal / DAKO Reações de imunoistoquímica realizadas com os kits de detecção Ventana Ultraview Universal DAB e/ou Ventana Optiview Universal DAB e/ou Sistema EnvisionFlex HRP-Dako. INTERPRETAÇÃO - O perfil imunoistoquímico, associado aos achados morfológicos, é compatível com o diagnóstico de PA) Biópsia percutânea à vácuo (mamotomia) de mama direita - QIM, distorção arquitetural, guiada por estereotaxia: Parênquima mamário com: - Alterações fibrocísticas. - Adenose simples. - Alterações colunares dos lóbulos. - Atividade secretora acinar. - Hiperplasia pseudo-angiomatosa estromal discreta. - Hiperplasia epitelial ductal usual moderada sem atipias. - Hiperplasia lobular atípica (0,8 mm; 2% da extensão total dos fragmentos). - Infiltração adiposa do estroma. PB) Biópsia percutânea à vácuo (mamotomia) de mama direita - QIM, calcificações, guiada por estereotaxia: - Carcinoma ductal "in situ" de grau intermediário (ver nota). Grau nuclear: 3 Padrão celular: usual (SOE) e apócrino Padrão arquitetural: sólido Comedonecrose: ausente Extensão para ácinos adjacentes do tecido mamário: presente Calcificações: presentes Maior dimensão da neoplasia nesta amostra: 2,9 mm (20% da extensão total dos fragmentos). Nota: O diagnóstico final depende da análise histológica completa da lesão. Não foi caracterizada invasão do estroma mesmo com o auxílio do exame imunoistoquímico, nos múltiplos cortes avaliados. Correlacionar com os exames de imagem. Análise morfológica do material realizada por patologia digital através do sistema de escaneamento de lâminas Aperio GT450 - Leica Biosystems. Controles internos presentes e adequados nos cortes das reações dos receptores hormonais (estrógeno e progesterona). Her-2 (CERBB2) - Interpretação (segundo critérios do CAP - College of American Pathologists) Detecção por método imunoistoquímico da oncoproteína HER-2. A reação pode sofrer a interferência do processo de fixação prévia. Estes testes não foram validados para material submetido a descalcificação. Valores de referência: ESCORE 0 (STATUS DE HER2 - NEGATIVO) - Ausência de imunomarcação, ou imunoreação fraca, incompleta, ou fraca / completa detectada em menos de 10% das células neoplásicas. ESCORE 1+ - Imunoreação fraca, incompleta, detectada em mais de 10% das células neoplásicas. ESCORE 2+ (STATUS DE HER2 - INCONCLUSIVO / EQUÍVOCO) - Imunomarcação fraca a moderada, não uniforme e incompleta, desenhando todo o contorno da membrana em mais de 10% das células neoplásicas; ou forte e completa em menos de 10% das células neoplásicas. ESCORE 3+ (STATUS DE HER2 - POSITIVO) - Imunomarcação forte e completa, desenhando todo o contorno da membrana em mais de 10% das células neoplásicas. Referência: Wolff AC, Somerfield MR, Dowsett M, et al. Human Epidermal Growth Factor Receptor 2 Testing in Breast Cancer: ASCO-College of American Pathologists Guideline Update. J Clin Oncol. 2023: JCO2202864. doi: 10.1200/JCO.22.02864.
"""
print(extrai_status_her2(TEXTO3))
print()
conta_siglas(TEXTO3)
```
